```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
md_rem = np.array([0.55, 0.49, 0.35, 0.59, 0.50, 0.25, 0.23, 0.11, 0.16, 0.41, 0.15, 0.08, 0.14, 0.43, 0.14, 0.25, 0.52, 0.66, 0.47, 0.33, 0.62, 0.56, 0.26, 0.69, 0.15, 0.58, 0.46, 0.25, 0.47, 0.05, 0.08, 0.22, 0.55, 0.26, 0.62, 0.28, 0.35, 0.63, 0.19, 0.31, 0.33, 0.40, 0.11, 0.21, 0.42, 0.44, 0.48, 0.51, 0.13, 0.5])
qtd_zero = np.zeros((len(md_rem), 1), dtype=int)
```

```python
dados_radiacao = pd.DataFrame(md_rem, columns=['Radiacao'])
dados_radiacao['qtd'] = qtd_zero
df = dados_radiacao.groupby('Radiacao').agg({'qtd': 'count'}).reset_index()
df.head()
```

Construa um distribuição de frequencia usando intervalos 0 - 0.10., 0.10 - 0.20 ... 0.90 - 1.00

```python
sns.set_style("darkgrid")
ax = sns.barplot(x="Radiacao", y="qtd", data=df, units= 10)
ax.set_title('Frequencia de Radiacao')
ax.set_xlabel('Radiacao')
ax.set_ylabel('Frequencia')
plt.show()
```

```python
sns.set_style("darkgrid")
plt.figure(figsize=(10, 6))
ax = plt.hist(md_rem, bins=10)
plt.title('Histograma de Frequência de Radiação')
plt.xlabel('Frequência', fontsize=12)
plt.ylabel('Nível de Radiação', fontsize=12)
plt.show()
```
